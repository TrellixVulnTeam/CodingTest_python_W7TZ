```python
from collections import defaultdict
import sys 
sys.setrecursionlimit(100000)

def solution(n, s, a, b, fares):
    answer = 0
    fare = defaultdict(list)
    for i in fares:
        c,d,f = i
        fare[c]+=[[d,f]]
        fare[d]+=[[c,f]]
#         fare[c].sort(key=lambda x:x[1])
#         fare[d].sort(key=lambda x:x[1])
        
    res = []
#     dist = []
    
    # 합승X
    # a<->s, b<->s사이의 최단비용 거리 구하기
    
    def dfs1(start,cost,r,v,costa):
        if start==r:
            costa.append(cost)
            
        if fare.get(start):
            for i in fare[start]:
                if i not in v:
                    dfs1(i[0],cost+i[1],r,v+[i],costa)
                    
    def dfs2(start,cost,r,v,costb):
        if start==r:
            costb.append(cost)
        
        if fare.get(start):
            for i in fare[start]:
                if i not in v:
                    dfs2(i[0],cost+i[1],r,v+[i],costb)
        
    costa = []
    visit=[s]        
    dfs1(s,0,a,visit,costa)  # 합승하지 않고 start ~ a까지의 최단 비용 거리
    
    costb = []
    visit=[s]
    dfs2(s,0,b,visit,costb)  # 합승하지 않고 start ~ b까지의 최단 비용 거리
    
    #print('costa : ',costa)
    #print('costb : ',costb)
    res.append(sorted(costa)[0]+sorted(costb)[0])
    
    #print('res : ',res)
        
        
    #합승O
    tmp = [x for x in list(range(1,n+1)) if x not in [a]]
    
    dista=defaultdict(int)
    for i in tmp:
        tmpa=sys.maxsize
        #visita=[0]*(n+1)
        #visita[i]=1    
        visita=[i]
        tmpas = []
        
        def dfsa(start,cost,visita):
            #print('a','start : ',start,'cost : ',cost)
            if start == a:
                tmpas.append(cost)
                     
            if fare.get(start):   
                for j in fare[start]:
                    if j[0] not in visita and j[0] !=s:
                        dfsa(j[0],cost+j[1],visita+[j[0]])
                        
        dfsa(i,0,visita)
        if tmpas:
            dista[i]=sorted(tmpas)[0]
        
    tmp = [x for x in list(range(1,n+1)) if x not in [b]]
    distb=defaultdict(int)
    for i in tmp:
        tmpb=sys.maxsize
        #visitb=[0]*(n+1)
        visitb=[i]
        tmpbs=[]
        def dfsb(start,cost,visitb):
            if start==b:
                tmpbs.append(cost)
                
            if fare.get(start):   
                for j in fare[start]:
                    if j[0] not in visitb and j[0] !=s:
                        visitb.append(j[0])
                        dfsb(j[0],cost+j[1],visitb+[j[0]])
        
        dfsb(i,0,visitb)
        if tmpbs:
            distb[i]=sorted(tmpbs)[0]
        
#     print('dista : ',dista)
#     print('distb : ', distb)

    tmp = [x for x in list(range(1,n+1)) if x not in [s]]
    distt=defaultdict(int)
    for i in tmp:
        tmpt=sys.maxsize
        visitt=[i]
        tmpts=[]
        def dfst(start,cost,visitt):
            if start==s:
                tmpts.append(cost)
                
            if fare.get(start):   
                for j in fare[start]:
                    if j[0] not in visitt:
                        visitt.append(j[0])
                        dfst(j[0],cost+j[1],visitt+[j[0]])
        
        dfst(i,0,visitt)
        if tmpts:
            distt[i]=sorted(tmpts)[0]
        
    #print(distt)
    
    result = sys.maxsize
    
    for k,v in distt.items():
        result+=v
        
    for i in tmp:
        if dista.get(i) and distb.get(i) and distt.get(i):
            result = min(result,dista[i]+distb[i]+distt[i])
    final = min(result,min(res))
    
    
    sub2 = sys.maxsize
    if dista.get(b) and distb.get(s):
        sub2 = dista[b]+distb[s]  

    final = min(final,sub2)
    #print(final)
    return final
```
