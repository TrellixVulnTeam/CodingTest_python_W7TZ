## 내 답 : 

```python
from collections import defaultdict
def solution(rows, columns, queries):
    answer = []
    
    x=1
    mat = []
    for i in range(rows):
        sub = []
        for j in range(columns):
            sub.append(x)
            x+=1
        mat.append(sub)
        
    for q in queries:
        d = defaultdict(int)
        x_start = q[0]-1
        y_start = q[1]-1
        x_end = q[2]-1
        y_end = q[3]-1
        sub_x = -1
        sub_y = -1    

        # 테두리 값 추출
        # 오른   
        for i in range(y_start,y_end+1):
            d[(x_start,i)] = mat[x_start][i]
            sub_x = x_start
            sub_y = i
            #print(d)

        # 아래
        for i in range(sub_x,x_end+1):
            d[(i,sub_y)] = mat[i][sub_y]
            sub_x = i
            sub_y = sub_y

        # 왼
        for i in range(sub_y,y_start-1,-1):
            d[(sub_x,i)] = mat[sub_x][i]
            sub_x = sub_x
            sub_y = i

        #위
        for i in range(sub_x,x_start-1,-1):
            d[(i,sub_y)] = mat[i][sub_y]
            sub_x = i
            sub_y = sub_y

        # 가장 작은 값을 answer에 넣음
        ans = min(list(d.values()))
        answer.append(ans)


        # 회전후 배열에 적용
        vec = list(d.values())
        last = vec.pop(-1)

        flag = True
        for k,v in d.items():
            if flag:
                d[k] = last
                flag = False
            else:
                d[k] = vec.pop(0)

                
        for k,v in d.items():
            mat[k[0]][k[1]] = v
        
    
    return answer
```
