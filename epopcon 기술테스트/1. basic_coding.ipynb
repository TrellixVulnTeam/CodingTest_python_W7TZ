# 답안

````python
import pandas as pd
import numpy as np
import names
import random 
import math

rows = 253
max_size = 10

def adult_check(df):
    '''
    주어진 df에 대해서 adult column을 추가하고 나이순으로 정렬해서 리턴하는 function.
    한번에 10개 까지만 처리 가능한 것으로 가정함.
    주의 : 이 function은 수정하면 안됨!!
    '''

    if len(df) > max_size:
        raise Exception('data frame too big to process')

    df.loc[df.index, 'adult'] = df.age >= 19

    return df.sort_values('age')


print('data frame input')
df = pd.DataFrame(
    {
        'name' : [ names.get_full_name() for i in range(rows) ], 
        'age' : [ random.randint(0, 59) for i in range(rows) ]
    }
)
print(df)

print('\ndata frame split')

df_split = np.array_split(df, (len(df)//9))     # 문제1. answer code here  (1 line으로 작성하면 더 좋음)
print(df_split[0])
print(df_split[-1])

print('\ndata frame result')
result = pd.concat([adult_check(x) for x in df_split]).sort_index(ascending=True)  # 문제2. answer code here  (1 line으로 작성하면 더 좋음)
print(result)
````

````python
import re

def dummy_tokenizer(string, split_by):
    # implement function 
    pattern = '|'.join(map(re.escape, split_by))   # answer code here (should be 1 line)
    return re.sub(pattern,' ', string).split()          # answer code here (should be 1 line)

def dummy_ngram(list_of_token, n):
    # implement function 
    return list(zip(*[list_of_token[i:] for i in range(n)]))     # answer code here (should be 1 line)



split_by = ['[', ']', ':', '+', ' ']
goods_name = '[[패키지 : 삼성 공기청정기]] [화이트 그랑데AI 패키지] 삼성 그랑데 AI 세탁기 21KG [WF21T6500KW] + 삼성 그랑데 AI 건조기 16KG [DV16T8740B'

list_of_token = dummy_tokenizer(goods_name, split_by)
n_gram = dummy_ngram(list_of_token, 7)

print(f'\ngoods_name : \n "{goods_name}"')
print(f'\nlist_of_token : \n {list_of_token}')
print('\nn gram :')
[print(g) for g in n_gram]
````
