```python
# DFS - 시간초과 
import sys
def solution(board):
    answer = 0
    cost = [sys.maxsize]
    
    def dfs(i,j,road,corner,v,d):
        if i>=row or i<0 or j>=col or j<0 or board[i][j]==1 or v[i][j]==1:
            return
        
        if i==row-1 and j==col-1:
            cost[0] = min(cost[0],road*100+corner*500)
            return

        visits=[[0]*col for _ in range(row)]
        for x in range(row):
            for y in range(col):
                visits[x][y]=v[x][y]
        visits[i][j]=1

        if d==0 or d==2: # 상 or 하
            dfs(i+1,j,road+1,corner,visits,2)  # 하
            dfs(i-1,j,road+1,corner,visits,0)   # 상
            dfs(i,j+1,road+1,corner+1,visits,3)  # 우
            dfs(i,j-1,road+1,corner+1,visits,1)   # 좌

        elif d==1 or d==3: # 좌 or 우
            dfs(i,j+1,road+1,corner,visits,3)  # 우
            dfs(i,j-1,road+1,corner,visits,1)   # 좌
            dfs(i+1,j,road+1,corner+1,visits,2)  # 하
            dfs(i-1,j,road+1,corner+1,visits,0)   # 상
            
        elif d==-1:
            dfs(i+1,j,road+1,corner,visits,2)  # 하
            dfs(i,j+1,road+1,corner,visits,3)  # 우
        return      
    
    row= len(board)    # 행의 길이
    col= len(board[0]) # 열의 길이
    visit = [[0]*col for _ in range(row)]
    dfs(0,0,0,0,visit,-1)
    
    return cost[0]
```

```python
board=[[0,0],[0,0]]
solution(board)
```
