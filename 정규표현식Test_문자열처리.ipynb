```python
import re

def solution(new_id):
    answer = ''
    # 1
    new_id = new_id.lower()
    #print(new_id)
    
    #2
    lst = ['-','_','.']
    ids = ''
    for i in new_id:
        if i in lst or i.isdigit() or i.isalpha():
            ids+=i
    new_id = ids
    

    #3
    cnt=0
    ids=''
    for i in new_id:
        if i=='.':
            cnt+=1
            
        else:
            if cnt>=1:
                ids+='.'
                cnt=0
            ids+=i
    new_id = ids
    
    #print(new_id)
    
    #4
    if len(new_id)>=1:
        if new_id[0]=='.':
            new_id = new_id[1:]
        if new_id[-1]=='.':
            new_id=new_id[:-1]
        
    #5
    if len(new_id) == 0:
        new_id+='a'
        
    #6
    if len(new_id)>=16:
        new_id = new_id[:15]
        if new_id[-1]=='.':
            new_id=new_id[:-1]
            
    #7
    if len(new_id)<=2:
        sub = new_id[-1]
        while(len(new_id)<3):
            new_id+=sub
    return new_id
    
    
    
    return answer
```
