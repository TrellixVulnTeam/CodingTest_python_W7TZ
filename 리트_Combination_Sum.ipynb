```python
from itertools import combinations                                  # 조합
from itertools import combinations_with_replacement  # 중복조합 
candidates=[2,3,6,7]
target=7

candidates=[2,3,5]
target=8

ans=[]
for i in range(1,target+1):
    lst=list(map(list,(combinations_with_replacement(candidates,i))))
    for j in lst:
        if sum(j)==target:
            ans.append(j)
            
ans
```

```python
import itertools
candidates=[5,10,8,4,3,12,9]
target=27
ans=[]
ans=[list(map(list,combinations_with_replacement(candidates,i))) for i in range(1,target+1) ]
answer=[i for i in itertools.chain(*ans) if sum(i)==target]   # itertools.chain : 리스트 unpacking 하는 역할
print(answer)
```
