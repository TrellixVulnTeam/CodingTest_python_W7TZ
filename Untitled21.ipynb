```python
from collections import defaultdict
def solution(table, languages, preference):
    answer = ''
    
    d_t = defaultdict(int)
    score = [5,4,3,2,1]
    d = {k:v for k,v in zip(languages,preference)}
    print(d)
    for i in table:
        
        t = i.split()
        
        tmp_d = {x:y for x,y in zip(t[1:],score)}
        print(tmp_d)
        s = 0
        for j in languages:
            if j in tmp_d:
                s += d[j]*tmp_d[j]  # 선호도 * 직군별 점수
        
        d_t[t[0]] = s
    print(d_t)
    return sorted(d_t.items(), key = lambda x: (-x[1],x[0]))

table = ["SI JAVA JAVASCRIPT SQL PYTHON C#", "CONTENTS JAVASCRIPT JAVA PYTHON SQL C++", "HARDWARE C C++ PYTHON JAVA JAVASCRIPT", "PORTAL JAVA JAVASCRIPT PYTHON KOTLIN PHP", "GAME C++ C# JAVASCRIPT C JAVA"]
languages = ["PYTHON", "C++", "SQL"]
preference = [7,5,5]
solution(table, languages, preference)
```

```python
S = '0000011'
print("S : ",S)
sub = len(S)
cnt = 0


while(1):
    
    if len(S)>1:
        lst = [0,0]
        f = S[0]
        for i in S[1:]:
            if f!=i:
                lst[int(f)] += 1
                f = i
        lst[int(f)]+=1
    
    print(lst)
    break
    
    if lst[1]>lst[0]:  # 0 -> 1로 바꿔
        
        start, end = -1, -1
        for i, j in enumerate(S):
            if j == '0' and start == -1:
                start = i
                if i == len(S) - 1:
                    end = i
                    break
                continue

            if j == '1' and start>=0:
                end = i - 1
                break

        tmp = '1'*(end-start+1)
        if end == len(S)-1:
            S = S[:start]+tmp
        else:
            S = S[:start]+tmp+S[end+1:]
    
    else:  # 1-> 0
        print("여기가 실행")
        start, end = -1, -1
        for i, j in enumerate(S):
            if j == '1' and start == -1:
                start = i
                continue
            if j=="1" and start>=0 and i==(len(S)-1):
                end = i
                break
                

            if j == '0' and start>=0:
                end = i-1
                break

        tmp = '0' * (end - start + 1)
        if end == len(S) - 1:
            S = S[:start] + tmp
        else:
            S = S[:start] + tmp + S[end + 1:]

    print("바꾼뒤 S : ",S)
    cnt+=1
    if S=='1'*sub or S=='0'*sub:
        break

print(cnt)
```
